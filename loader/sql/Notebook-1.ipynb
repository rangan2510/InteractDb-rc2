# Find interactions for Cyclosporin & Tacrolimus that are not present in DrugBank directly.

```sql
SELECT *
FROM drug_nodes
WHERE synonyms LIKE '%Cyclosporin%'
```

Here, we find that Cyclosporin has ATC =Â L04AD01. We will use this for reference from now on.

```sql
SELECT Drug1.name AS 'Source Drug', Drug2.name AS 'Interacting Drug', [dbo].[drug_drug_edges].[desc] AS 'Effect of Interaction'
FROM drug_nodes Drug1, drug_drug_edges, drug_nodes Drug2
WHERE MATCH(Drug1-(drug_drug_edges)->Drug2)
AND Drug1.atc = 'L04AD01'
```

Now, let's look at all the genes that are targeted by this drug

```sql
SELECT Drug1.name AS [Drug 1], Genes.gene_names, Genes.uniprotkb_id, drug_gene_edges.action
FROM drug_nodes Drug1, drug_gene_edges, gene_nodes Genes
WHERE MATCH(Drug1-(drug_gene_edges)->Genes)
AND Drug1.atc = 'L04AD01'
```

```sql
SELECT Drug1.name AS [Drug 1], dge1.action, Genes.gene_names AS targets, Genes.uniprotkb_id, dge2.action, Drug2.name AS [Drug 2]
FROM drug_nodes Drug1, drug_gene_edges dge1, gene_nodes Genes, drug_gene_edges dge2, drug_nodes Drug2
WHERE MATCH(Drug1-(dge1)->Genes<-(dge2)-Drug2)
AND Drug1.atc = 'L04AD01'
AND Drug2.atc != Drug1.atc
ORDER BY [targets] ASC
```
