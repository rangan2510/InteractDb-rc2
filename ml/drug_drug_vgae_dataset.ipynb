```python
import pandas as pd
```

```python
from rdkit.Chem import PandasTools
SDFFile = "../data/drugs/drugbank_5.1.10/structures.sdf"
struct_df = PandasTools.LoadSDF(SDFFile,strictParsing=False)
struct_df
```

```python
struct_df.columns
```

```python
df = struct_df[['DATABASE_ID', 'SMILES', 'MOLECULAR_WEIGHT', 'JCHEM_AVERAGE_POLARIZABILITY','JCHEM_BIOAVAILABILITY','JCHEM_DONOR_COUNT', 'JCHEM_FORMAL_CHARGE','JCHEM_NEUTRAL_CHARGE', 'JCHEM_NUMBER_OF_RINGS','JCHEM_PHYSIOLOGICAL_CHARGE', 'JCHEM_PKA', 'JCHEM_PKA_STRONGEST_ACIDIC','JCHEM_PKA_STRONGEST_BASIC', 'JCHEM_POLAR_SURFACE_AREA','JCHEM_REFRACTIVITY', 'JCHEM_ROTATABLE_BOND_COUNT']]
df.to_csv("nodes.tsv", sep="\t", index=False)
```

```python
drug_ids = df['DATABASE_ID'].to_list()
```

```python
int_df = pd.read_csv(r"..\parser\temp\drug-interactions.tsv", sep="\t")
int_df = int_df[["source","dst","desc"]]
int_df
```

```python
int_df = int_df[int_df.source.isin(drug_ids)]
int_df = int_df[int_df.dst.isin(drug_ids)]
int_df
```

```python
int_df = int_df[['source','dst']]
int_df.to_csv("interactions.tsv", sep="\t", index=False)
```

```python
cancer_df = pd.read_csv(r"D:\Github\Cancer-Drug-Interactions\nodes_mapped_from_db.tsv", sep='\t')
cancer_df
```
