```python
import pandas as pd
df = pd.read_excel(r"..\data\diseases\DB_to_ICD11\DB_to_ICD.xlsx")
df
```

```python
df_icd = pd.read_csv("./temp/icd11_10.tsv", sep="\t",)
df_icd
```

```python
df_cond_icd = pd.merge(df, df_icd, left_on="ICDCode", right_on="icd11Code", how="left")
```

```python
df_cond_icd = df_cond_icd[["drugbank-id", "assoc_cond", "icd11Code", "icd11Title", "icd10Code", "icd10Title"]]
df_cond_icd
```

```python
# df_cond_icd = df_cond_icd.dropna()
# df_cond_icd
```

```python
df_disgenet = pd.read_csv("./temp/disease_icd.csv", sep="\t")
df_disgenet
```

```python
df_disgenet = df_disgenet[df_disgenet['vocabulary'].str.contains('ICD10', na=False)]
```

```python
df_db_dis = pd.merge(left=df_cond_icd, right=df_disgenet, left_on="icd10Code", right_on="vocab_code", how="inner")
df_db_dis = df_db_dis[["drugbank-id", "disgenet_id","name", "assoc_cond", "icd11Code", "icd11Title", "icd10Code", "icd10Title"]]
df_db_dis
```

```python
df_db_dis.to_csv("./temp/drug_disease_edges.csv", sep="\t", index=False)
```
