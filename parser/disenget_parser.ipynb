# Gene (Uniprot) to Disease id mapping for edges

```python
import sqlite3

conn = sqlite3.connect('../data/diseases/DisGeNET/disgenet_2020.sqlite')
query = """
    SELECT DISTINCT * FROM geneAttributes
    JOIN geneDiseaseNetwork 
    ON geneAttributes.geneNID = geneDiseaseNetwork.geneNID
    JOIN diseaseAttributes
    ON diseaseAttributes.diseaseNID = geneDiseaseNetwork.diseaseNID
"""

disgen_df = pd.read_sql_query(query,conn)
disgen_df
```

```python
gene_map = pd.read_csv('../data/diseases/DisGeNET/mapa_geneid_4_uniprot_crossref.tsv', sep="\t")
gene_map.columns = ['uniprotkb_id','geneId']
gene_map
```

```python
disgen_df2 = pd.merge(left=disgen_df, right=gene_map, on='geneId', how='left')
```

```python
disgen_df2 = disgen_df2[['uniprotkb_id','diseaseId','geneName', 'geneDescription', 'pLI', 'DSI', 'DPI', 'source','associationType', 'sentence', 'pmid']]
disgen_df2.columns = ['src','dst','gene_name', 'gene_description', 'pLI', 'DSI', 'DPI', 'source','association_type', 'sentence', 'pmid']
disgen_df2
```

```python
disgen_df2 = disgen_df2.loc[disgen_df2['dst'].isin(disease_list)]
```

```python
disgen_df2.to_csv('./temp/disease_gene.csv', sep="\t", index=False)
```
